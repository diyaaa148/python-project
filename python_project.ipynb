EXPLORATORY DATA ANALYSIS

GENDER

```python
df.columns
```

```python
ax= sns.countplot(x= 'Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender', y = 'Amount' , data= sales_gen)
```

from the graph we can see that most of the buyer are fenales and even the purchasing power of females are greater than that of men

```python
df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

AGE

```python
ax= sns.countplot(data=df, x='Age Group',hue= 'Gender' )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount v/s age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount', data=sales_age)
```

from above graph we can see that most of the buyers are of the age group between 26-35 yrs female

STATE

```python
#total no. of orders from top10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State',y='Orders')
```

```python
# total column/ sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State',y='Amount')
```

from the above graph we can see that unexpectedly most of the orders are from UP , maharastra & karnataka respectively but sales are most from up, maharastra and karnataka

MARITAL STATUS

```python
ax= sns.countplot(data= df, x='Marital_Status', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

OCCUPATION

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='Occupation',y='Amount')
```

PRODUCT CATEGORY

```python
ax= sns.countplot(data= df, x='Product_Category')
x=sns.set(rc={'figure.figsize':(30,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data=sales_product, x='Product_Category',y='Amount')
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data=sales_product, x='Product_Category',y='Orders')
```

from the above graphs w can see that most of the products are from food , footwear and clothing
